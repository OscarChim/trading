<h2> Testing Consistency for Reversal Strategy </h2>

<h3> Idea </h3>

The idea of "testing the consistency for reversal strategy" is to validate if a group of etfs performs a strong (weak) reversal characteristic in the history, whether they will continue to perform a strong (weak) reversal characteristic in the future?  
  
In general, there are several steps as follows.
- group the etfs according to historical reversal performance (100 days)
- focus on the group with strongest/weakest reversal
- test the reversal performance via 3 different regression methods for the next 20 days
- calculate the regression coefficient every 20 days, at the same time, update the grouping
- compare the historical and future performance (t-statistics for regression coefficient)

<h3> Methods </h3>

<h4> Setting Groups </h4>
- for each etf (total 550 etfs),run time series regression from t to t+100 (100 days)
$$ R_{t+1} =  b_0 + b_1*R_t+ e_t $$
- If there is a strong reversal, then b1 must be smaller than zero (assume benchmark = 0)
- We get in total 550 b1
- group the eft according to the value of b1 into 10 groups
- group 0 has the smallest b1, and group 9 has the largest b1


<h4> Testing reversal for group 0 and group 9 </h4>
- Run the following regression (general idea formula) within a given group from t+101 to t+120 (20 days)
$$ R_{t+1}= a_0+a_1*R_t+a_2*MA(5)_t $$
- R_t means the return for today, MA(5) is the average return for the past 5 days (including today)
- R_t+1 means the return for the next day
- Take group 0 as an example, if reversal is consistent, then a low a1 or a2 (or combination effect) will follows a low b1
- we examine the t statistics for a1 and a2

<h4> Regression methods </h4>
  
We perform 3 different regression methods for testing reversal, which are basically derived from the general idea formula above. 
  
a. portfolio method
- let the 55 etf in one group forms a portfolio
- R_t and MA(5) are both simple average for the 55 etfs data at given date t
- this could examine the overall effect of reversal for the given group
  
b. time demeaned method
- since the testing data are a panel data with 55 etfs and 20 days, there might be time fixed effect, which will weaken the validity of the estimate of coefficients
- to eliminate the time fixed effect, we improve the regresson as follows
$$ R_{t+1} - 1/S * \sum_{i=1}^S R_{t+1} = a_0+a_1* [R_t - 1/S * \sum_{i=1}^S R_{t}]+a_2*[MA(5)_{t} - 1/S * \sum_{i=1}^S MA(5)_{t}] $$
- By subtracting the average value for 55 efts (here S = 55) in a given date, we can eliminate the time fixed effect
- we can also interpret this regression formula as how a particular etf outperform the benchmark (average value in the group) in reversal
- this regression runs 1100 rows of data in one time (55 * 20)  
   
c. Fama_Macbeth method
- In Fama method, for the 20 days testing data, we run regression on one day for 20 times (t+101 to t+120), each regression include 55 etfs data
- average the 20 estimates of coefficients using Fama_Macbeth technique
$$ a_{fama} = 1/T * \sum_{i=1}^T a_{t} $$
- it can improve the residual correlation problem in OLS

<h4> Rolling throughout whole dataframe </h4>
  
- get the coefficient estimates for every 20 days
- update the groups every 20 days
- for example: 
- t = 1 to t = 100: get the first 550 b1 and form the first 10 groups
- t = 101 to t = 120: get the first estimate of a1 and a2
- t = 20 to t = 120: get the second 550 b1 and form the second 10 groups
- t = 121 to t = 140: get the second estimate of a1 and a2
  
- in general, there are 139 estimates of a1 and a2 for the whole dataframe ( (2893 - 120)/20 +1 = 139.65)

<h2> Study on reversal in ETFs with different market capitalization</h2>

<h3> Idea </h3>

The idea of ‘study on reversal in ETFs with different market capitalization’ is to compute reversal performance with companies with different market capitalization and whether big companies and small ones have the same reversal effect？Whether the time period will affect the reversal performance?

In general, there are several steps as follows.

- Divide the 548 ETFs into 4 groups according to their market cap(from smallest to largest)
- Test the reversal performance for each group by regressing returns on lagged returns
- Compare coefficients and validate the significance( t-test)
- Further confirm a longer horizon reversal

<h3> Methods </h3>

<h4> Setting Groups </h4>

- Calculate lagged return and drop ETFs which has unavailable data, finally each date (for 2892 days) consist of exactly 548 ETF data
- Calculate market cap : $$MKTCAP = SHROUT* |PRC|$$
- Divide ETFs into 4 groups by market cap (137 ETFs/each group), smallest ones are 1, and largest ones are 4

<h4> Testing reversal for group 1 to 4 </h4>

- Making hypothesis:  
 H0: Large Market cap (top 25%) and Small Market cap (bottom 25%) ETFs have the same reversal effect.  
 H1: Large Market cap (top 25%) has a smaller reversal effect than the Small Market cap ETFs (bottom 25%).

- Run the following regression for different groups:
$$ R_{t+1} =  b_0 + b_1*R_t+ e_t $$
R_t means the return for today，R_t+1 means the return for the next day
- Examine the t statistics for b1 for each group

<h4> Testing difference between group 1 and group 4 </h4>

- Calculate difference lagret between smallest and lagest market cap group
- We examine the t statistics for new b1

<h4> Take a long term look for reversal performance in small cap  </h4>

- Pick weekly data from daily data by using the day of the week
- Calulate new lagged return and drop unavailable ones
- Divide ETFs into 4 groups by new market cap
- Run the following regression for different groups:
$$ R_{t+1} =  b_0 + b_1*R_t+ e_t $$
- Examine the t statistics for b1 for each group
- Calculate difference between smallest and lagest market cap group

<h4> Analyze a new portofolio </h4>

- Build a portfolio that only contains ETFs with a small cap
- Subdivision groups by using lagged return
- Calculate average return
- Sell ETFs with highest return and buy ETFs with lowest return
- Calculate mean, standard deviation and sharpe ratio

```python
get_date = pd.DataFrame(reg_data.groupby('DATE')['PERMNO'].count())
list_date = get_date.index.tolist()
len(list_date)
```

Now get the beta to further group the eths into 10 groups

```python
def regfun(df, mod):
    results = smf.ols(mod, data=df).fit()
    return results.params
```

```python
allparams = reg_data_1.groupby('PERMNO').apply(regfun, 'RET ~ LAGRET')
```

group the parameter into 10 groups

```python
allparams['QUINTILE'] = pd.qcut(allparams.LAGRET,10,labels=False)
allparams
```

get the first b1 for the lowest beta group

```python
group_0 = allparams.loc[allparams['QUINTILE'] == 0]
group_0_beta = group_0['LAGRET'].mean()
group_0_beta
```

get the first test dataset: etf lays in group 0 and for the 101 to 120 dates

```python
group_0_etf_list = group_0.index.to_list()
list_date_test_1 = list_date[100:120]
test_data_1 = reg_data[reg_data['DATE'].isin(list_date_test_1) & reg_data['PERMNO'].isin(group_0_etf_list) ]
test_data_1
```

aggregate the mean for the 55 etfs in a given day

```python
try1 = test_data_1.groupby('DATE').agg('mean')
try1
```

run the first test data regression to get the estimate of a0 and a1

```python
test_reg_result = regfun(try1, 'RET ~ LAGRET + MA5')
```

```python
test_reg_result
```

<h3> Loop 1: Portfolio method, strongest reversal group </h3>

```python
df_compare_beta = pd.DataFrame(columns= ['Current_Date', 'beta_RET_1_100',
                                         'intercept_1_100','beta_RET_101_120','beta_MA5_101_120','intercept_101_120'])
```

```python
for i in range(0,len(list_date)-120,20):
    list_date_i = list_date[i:i+100]
    reg_data_i = reg_data[reg_data['DATE'].isin(list_date_i)]
    allparams_i = reg_data_i.groupby('PERMNO').apply(regfun, 'RET ~ LAGRET')
    allparams_i['QUINTILE'] = pd.qcut(allparams_i.LAGRET,10,labels=False)
    group_0_i = allparams_i.loc[allparams_i['QUINTILE'] == 0]
    group_0_i_beta = group_0_i['LAGRET'].mean()   #b1
    group_0_i_intercept = group_0_i['Intercept'].mean()  #b0
    group_0_i_etf_list = group_0_i.index.to_list()
    list_date_test_i = list_date[i+100:i+120]
    test_data_i = reg_data[reg_data['DATE'].isin(list_date_test_i) & reg_data['PERMNO'].isin(group_0_i_etf_list) ]
    df_reg_test_i = test_data_i.groupby('DATE').agg('mean')
    test_reg_result_i = regfun(df_reg_test_i, 'RET ~ LAGRET + MA5')
    group_0_i_test_beta_1 = test_reg_result_i['LAGRET']  # a1
    group_0_i_test_beta_2 = test_reg_result_i['MA5']     # a2
    group_0_i_test_intercept = test_reg_result_i['Intercept']  #a0
    current_date = list_date[i+120]
    df_compare_beta = df_compare_beta.append({'Current_Date' : current_date, 'beta_RET_1_100' : group_0_i_beta, 
                                              'intercept_1_100' : group_0_i_intercept, 'beta_RET_101_120' : group_0_i_test_beta_1,
                                              'beta_MA5_101_120' : group_0_i_test_beta_2, 'intercept_101_120' : group_0_i_test_intercept}, 
                ignore_index = True)

df_compare_beta 
```

<h2> Portfolio by reversal strategy </h2>

According to the previous analysis, we can build a portfolio as follows:
- since the group with strongest reversal has significant reversal effect for future returns, we focus on group 0 (strongest reversal)
- There are 55 etfs in group 0
- According to the return of previous day, group them into 5 category (Q0:lowest ret; Q4:highest ret)
- __Strategy: Long the etfs in Q0 and short the assets in Q4__
- if reversal do exist, we can earn money from this strategy


```python
port_return_list = []  # this is the return list for the portfolio given above

for i in range(0,len(list_date)-120,20):
    list_date_i = list_date[i:i+100]
    reg_data_i = reg_data[reg_data['DATE'].isin(list_date_i)]
    allparams_i = reg_data_i.groupby('PERMNO').apply(regfun, 'RET ~ LAGRET')
    allparams_i['QUINTILE'] = pd.qcut(allparams_i.LAGRET,10,labels=False)
    group_0_i = allparams_i.loc[allparams_i['QUINTILE'] == 0]
    group_0_i_beta = group_0_i['LAGRET'].mean()   #b1
    group_0_i_intercept = group_0_i['Intercept'].mean()  #b0
    group_0_i_etf_list = group_0_i.index.to_list()
    list_date_test_i = list_date[i+100:i+120]
    test_data_i = reg_data[reg_data['DATE'].isin(list_date_test_i) & reg_data['PERMNO'].isin(group_0_i_etf_list) ]
    
    for j in range(i+100,i+120):
        list_date_test_j = list_date[j]
        test_data_j = test_data_i.loc[test_data_i['DATE']==list_date_test_j]
        test_data_j['QUINTILE'] = pd.qcut(test_data_j.LAGRET,5,labels=False)
        agg_result = test_data_j.groupby('QUINTILE').agg('mean')
        port_return = agg_result.loc[0,'RET'] - agg_result.loc[4,'RET']
        port_return_list.append(port_return)
```

As usual, we will group the ETFs each day according to their last day's return, we are going to long the lowest return group and short the highest return group.
